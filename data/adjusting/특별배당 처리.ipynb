### TROW

```python
TROW = pd.read_csv('Data_Dividend/TROW_d.csv')

TROW
```

```python
TROW2 = TROW.copy()

TROW2['Date'] = pd.to_datetime(TROW2['Date'])

# 2014년 2월 1일부터 2024년 3월 1일까지의 날짜 필터링
start_date = '2014-03-12'
end_date = '2023-12-14'
filtered_TROW = TROW2[(TROW2['Date'] >= start_date) & (TROW2['Date'] <= end_date)]

# 결과 출력
filtered_TROW
```

```python
data = filtered_TROW

index = list(range(110, 152))
filtered_TROW = pd.DataFrame(data, index=index)

# 제거할 인덱스
indices_to_remove = [115, 141]

# 특정 인덱스의 행 제거
filtered_TROW_revised = filtered_TROW.drop(indices_to_remove)

# 데이터프레임을 CSV 파일로 저장 (인덱스를 포함하지 않음)
filtered_TROW_revised.to_csv('TROW_d_revised.csv', index=False)
```
