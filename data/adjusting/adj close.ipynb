```python

def fetch_adjusted_close(stock_symbol, start_date, end_date):
    """
    Yahoo Finance에서 주어진 기간 동안 특정 주식의 조정 종가 데이터 import
    
    Args:
    stock_symbol (str): ticker
    start_date (str): start date ('YYYY-MM-DD')
    end_date (str): end date ('YYYY-MM-DD')
    
    Returns:
    pandas.DataFrame: 조정 종가 데이터를 포함하는 DataFrame
    """
    
    # 주식 데이터 import
    stock_data = yf.Ticker(stock_symbol)
    
    # 지정된 기간 동안의 주식 가격 데이터 import
    hist = stock_data.history(start=start_date, end=end_date)
    
    # return adj close
    return hist['Close']
```

```python
import os
import yfinance as yf

def fetch_adjusted_close_from_folder(folder_path, start_date, end_date):
    """
    Yahoo Finance에서 주어진 기간 동안 특정 주식의 조정 종가 데이터 import
    
    Args:
    stock_symbol (str): ticker
    start_date (str): start date ('YYYY-MM-DD')
    end_date (str): end date ('YYYY-MM-DD')
    
    Returns:
    pandas.DataFrame: 조정 종가 데이터를 포함하는 DataFrame
    """
    # 폴더 내의 모든 파일 목록 import
    ticker_files = [f for f in os.listdir(folder_path) if f.endswith('_d.csv')]
    
    # 각 파일명에서 티커명 추출
    tickers = [filename.split('_')[0] for filename in ticker_files]
        # 각 티커에 대한 adj close data를 딕셔너리에 저장
    adjusted_closes = {}
    for ticker in tickers:
        stock_data = yf.Ticker(ticker)
        hist = stock_data.history(start=start_date, end=end_date)
        adjusted_closes[ticker] = hist['Close']
    
    return adjusted_closes


folder_path = '/Users/justhaji/Desktop/mu-ah/adv 24/FE/Y-FoRM/장플/24 봄 장플/Data_Dividend'
start_date = '2014-02-01'
end_date = '2024-02-29'
ticker_data = fetch_adjusted_close_from_folder(folder_path, start_date, end_date)
```

```python
import pandas as pd

df = pd.DataFrame(ticker_data)
```

```python
df
```

```python
df.index = pd.to_datetime(df.index)

# 인덱스에서 시간 부분 제거
df.index = df.index.date

df
```

```python
# 인덱스를 'Date'라는 이름의 열로 변환
df.reset_index(inplace=True)
df.rename(columns={'index': 'Date'}, inplace=True)

df
```

```python
sorted_columns = sorted([col for col in df.columns if col != 'Date'])
df = df[['Date'] + sorted_columns]

df
```

```python
# 결과 데이터프레임 저장
output_path = 'adj close.csv'
df.to_csv(output_path, index=False)
```
